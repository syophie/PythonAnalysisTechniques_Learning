```python
import pandas as pd
import statsmodels.formula.api as sm

# dummy data:
df = pd.DataFrame({"A": [10,20,30,40,50], 
                   "B": [20, 30, 10, 40, 50], 
                   "C": [32, 234, 23, 23, 42523]})
```

```python
#OLS
result = sm.ols(formula="A ~ B + C", data=df).fit()
print(result.params)
# Coefficients:
#   Intercept = 14.952480 - Not significant i.e., Not statistically different from zero / not meaningful.
#   B = 0.401182 P=0.6000 - Not significant i.e., No reliable directional effect.
#   C = 0.000352 p=0.583 + Confidnece Interval includes zero - Not significant i.e., No reliable evidence that C predicts A.

print(result.summary())
# Model fit:
#   P-Value suggests not statistically significant
#   R vs Adj R suggest overfitting -> B and C do not explain A and can be dropped

# Standard Errors and t-tests:
#   All coefficients have classic signs that the model has too few observations and no real signal is being detected:
#       Large standard errors
#       Low tâ€‘values
```
